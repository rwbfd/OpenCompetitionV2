```python
from open_competition.tabular.encoder import *
from sklearn.datasets import make_classification
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X, y = make_classification() # create a fake dataset
x_col = list(map(str, list(range(20))))
df = pd.DataFrame(X, columns=x_col)
df_ = df.copy(deep=True)
df_['label'] = y
Y = 'label'
targets = x_col[:10]
targets_ = [x_col[:10], x_col[10:]]
```

DimReducEncoder

```python
dre = DimReducEncoder()
```

```python
config = [('pca', {'n_components': 5})]
targets_ = [x_col[:10], x_col[10:]]
dre.fit(df, targets_, config)
dre.transform(df)
```

```python
config = [('tsne', {'n_components': 3})]
targets_ = [x_col[:10], x_col[10:]]
dre.fit(df, targets_, config)
dre.transform(df)
```
